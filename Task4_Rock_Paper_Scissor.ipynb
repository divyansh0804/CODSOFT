```python
import random
count_rock=0
count_scissor=0
count_paper=0
def update_counts(user_input):
  global count_rock,count_paper,count_scissor
  if user_input==0:
    count_rock+=1
  elif user_input==1:
    count_paper+=1
  else:
    count_scissor+=1
def predict():
  if count_rock > count_paper and count_rock > count_scissor:
    i=0
  elif count_paper > count_rock and count_paper > count_scissor:
    i=1
  elif count_scissor> count_rock and count_scissor > count_paper:
    i=2
  else:
    i = random.randint(0, 2)
  return i
player_score=0
computer_score=0
def update_scores(user_input):
  global player_score, computer_score
  prediction=predict()
  if user_input==0:
    if prediction==0:
      print("You played ROCK, computer played ROCK.")
      print("Computer Score:", computer_score, "\nYour Score: ", player_score)
    elif prediction==1:
      print("You played ROCK, computer played PAPER.")
      computer_score+=1
      print("Computer Score: ", computer_score, "\nYour Score: ", player_score)
    else:
      print("You played ROCK, computer played SCISSORS.")
      player_score+=1
      print("Computer Score: ", computer_score, "\nYour Score: ", player_score)
  elif user_input==1:
    if prediction==1:
      print("You played PAPER, computer played PAPER.")
      print("Computer Score: ", computer_score, "\nYour Score: ", player_score)
    elif prediction==0:
      print("You played PAPER, computer played ROCK.")
      player_score+=1
      print("Computer Score: ", computer_score, "\nYour Score: ", player_score)
    else:
      print("You played PAPER, computer played SCISSORS.")
      computer_score+=1
      print("Computer Score: ", computer_score, "\nYour Score: ", player_score)
  else:
    if prediction==2:
      print("You played SCISSORS, computer played SCISSORS.")
      print("Computer Score: ", computer_score, "\nYour Score: ", player_score)
    elif prediction==1:
      print("You played SCISSORS, computer played PAPER.")
      player_score+=1
      print("Computer Score: ", computer_score, "\nYour Score: ", player_score)
    else:
      print("You played SCISSORS, computer played ROCK.")
      computer_score+=1
      print("Computer Score: ", computer_score, "\nYour Score: ", player_score)
valid_entries = ['0', '1', '2']
while True:
  user_input = input("Enter 0 for ROCK, 1 for PAPER and 2 for SCISSORS: ")
  while user_input not in valid_entries:
    print("Invalid Input!")
    user_input = input("Please!  Enter 0 for ROCK, 1 for PAPER and 2 for SCISSORS: ")
  user_input = int(user_input)
  update_scores(user_input)
  update_counts(user_input)
  print("")
  if computer_score==10:
    print("Computer Won!")
    break
  elif player_score==10:
    print("You Won!")
    break
```
